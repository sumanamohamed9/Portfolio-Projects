```r
#Load Packages
library(tidyverse)
library(dplyr)
library(geosphere)
```

```r
#importing dataset
data1 <- read_csv("/kaggle/input/cyclistic-bike-share/202201-divvy-tripdata.csv")
data2 <- read_csv("/kaggle/input/cyclistic-bike-share/202202-divvy-tripdata.csv")
data3 <- read_csv("/kaggle/input/cyclistic-bike-share/202203-divvy-tripdata.csv")
data4 <- read_csv("/kaggle/input/cyclistic-bike-share/202104-divvy-tripdata.csv")
data5 <- read_csv("/kaggle/input/cyclistic-bike-share/202105-divvy-tripdata.csv")
data6 <- read_csv("/kaggle/input/cyclistic-bike-share/202106-divvy-tripdata.csv")
data7 <- read_csv("/kaggle/input/cyclistic-bike-share/202107-divvy-tripdata.csv")
data8 <- read_csv("/kaggle/input/cyclistic-bike-share/202108-divvy-tripdata.csv")
data9 <- read_csv("/kaggle/input/cyclistic-bike-share/202109-divvy-tripdata.csv")
data10 <- read_csv("/kaggle/input/cyclistic-bike-share/202110-divvy-tripdata.csv")
data11 <- read_csv("/kaggle/input/cyclistic-bike-share/202111-divvy-tripdata.csv")
data12 <- read_csv("/kaggle/input/cyclistic-bike-share/202112-divvy-tripdata.csv")
```

```r
#quick review of te data structure and heading
glimpse(data12)
```

```r
trip <- bind_rows(data1,data2,data3,data4,data5,data6,data7,data8,data9,data10,data11,data12)
```

```r
# checking missing values
# this counts missing value (NA)for each column in the trips dataset
# the output helps determine which column has NAs and how many
colSums(is.na(trip))

# remove rows with missing values
# delete any rows containing the NA values
# this also ensure the dataset is cleaned but might reduce data size significantly
 trip <- trip %>% drop_na() 
```

```r
#creating month column
trip <- trip%>% mutate(month=format(as.Date(started_at),"%b"))


#creating year, day of week and ended hour
trip <- trip%>% mutate(year=format(as.Date(started_at), "%Y"),
                      day_of_week=format(as.Date(started_at), "%A"),
                       hour=format(as.POSIXct(ended_at), "%H")
                       )
#creating ride length in minutes
trip$ride_length <- (as.double(difftime(trip$ended_at, trip$started_at)))/60

#creating ride ditance in km
trip$ride_distance <- distGeo(matrix(c(trip$start_lng, trip$start_lat),ncol= 2),
                             matrix(c(trip$end_lng, trip$end_lat),ncol= 2)
                             )
# Ordering day_of_week column
trip$day_of_week <- ordered(trip$day_of_week, levels = c("Monday", "Tuesday", "Wednesday", 
                                                            "Thursday", "Friday", "Saturday", 
                                                            "Sunday"))

# Ordering month column
trip$month <- ordered(trip$month, levels = c("Jan", "Feb", "Mar", "Apr", "May", "Jun", 
                                                "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"))

#renaming columns
names(trip)[2] <- 'bike'
names(trip)[13] <- 'user'

#removing rows with NA
trip <- drop_na(trip)

#remove ride where values is negative and above 1 day
trip <- trip[trip$ride_length >= 1 & trip$ride_length <= 1440, ]
```

```r
# count the number of rides by casual and member riders
usercount <- trip %>%
group_by(user) %>%
summarize(total = n()) %>%
mutate(overall_total = sum(total)) %>%
group_by(user) %>%
summarize(percent_total = total / overall_total)

usercount
```

```r
#analyse the bike type used
bike_usage <- trip %>%
  group_by(bike) %>%
  summarize(total = n()) %>%
  mutate(overall_total = sum(total)) %>%
  group_by(bike) %>%
  summarize(percent_total =(total / overall_total))

bike_usage
```

```r
#analyse the bike type used by casual members
annual_memeber_biketype_used <- trip %>%
   filter(user== "member")%>%
   group_by(bike) %>%
  summarize(total = n()) %>%
  mutate(overall_total = sum(total)) %>%
  group_by(bike) %>%
  summarize(percent_casual =(total / overall_total))

annual_memeber_biketype_used
```

```r
#average ride duration by users
rider_time_user <- trip%>%
  group_by(user) %>%
  summarize(avg_time= mean(ride_length), .groups = "drop")

rider_time_user$avg_time <- round(rider_time_user$avg_time, digits = 0)

rider_time_user
```

```r
#average time duration by different user and bike type
rider_time_biketype <- trip%>%
  group_by(bike, user) %>%
  summarize(avg_time= mean(ride_length), .groups = "drop")

rider_time_biketype$avg_time <- round(rider_time_biketype$avg_time, digits = 0)

rider_time_biketype
```

```r
#average ride distance by user in KM
rider_dist_user <- trip%>%
  group_by(user) %>%
  summarize(avg_dist= mean(ride_distance), .groups = "drop")

rider_dist_user$avg_dist <- round(rider_dist_user$avg_dist, digit = 2)

rider_dist_user
```

```r
#average ride distance by different user and bike type in KM
rider_dist_biketype <- trip%>%
  group_by(bike, user) %>%
  summarize(avg_dist= mean(ride_distance), .groups = "drop")

rider_dist_biketype$avg_dist <- round(rider_dist_biketype$avg_dist, digit = 2)

rider_dist_biketype
```

```r
#hours spend by users for riding
hour <- trip %>%
  group_by(user, hour) %>%
  summarize(numberofrides = n(), .groups = 'drop')%>%
  arrange(hour)

hour
```

```r
#average ride distance by user
avgridedist <- trip%>%
  group_by(user) %>%
  summarize(avgdist = mean(ride_distance))

avgridedist$avgdist <- round(avgridedist$avgdist, digits = 3)

avgridedist
```

```r
#analyse the ride pattenr by user and day of week
dayofweek <- trip %>%
group_by(user, day_of_week) %>%
  summarize(numberofrides = n(), .groups = 'drop')

dayofweek
```

```r
#monthly ride trend by user
month <- trip %>%
group_by(user, month) %>%
  summarize(numberofrides = n(), .groups = 'drop')

month
```
